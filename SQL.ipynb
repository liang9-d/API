```python
import sqlite3

conn = sqlite3.connect(":memory:")  # local in-memory DB
df_drop.to_sql("store", conn, index=False, if_exists="replace")
```

# Top 10 store with Low prices:

```python
query = """
SELECT
  name,
  city,
  rating,
  review_count,
  score,

  RANK() OVER (
    ORDER BY score DESC, review_count DESC
  ) AS price_range_rank
FROM store
WHERE price_level = 'Low'
ORDER BY price_range_rank
LIMIT 10;


"""
pd.read_sql(query, conn)
```

# Ranking for city:

```python
query = """
WITH ranked AS (
  SELECT
    name,
    city,
    rating,
    review_count,
    score,
    RANK() OVER (
      PARTITION BY city
      ORDER BY score DESC
    ) AS city_rank,
    COUNT(*) OVER (
      PARTITION BY city
    ) AS total_stores_in_city
  FROM store
)
SELECT
  name,
  city,
  rating,
  review_count,
  score,
  city_rank,
  total_stores_in_city,
  ROUND(100.0 * city_rank / total_stores_in_city, 2) AS percent_rank
FROM ranked
WHERE name LIKE '%ICI tea drink%';

"""
pd.read_sql(query, conn)
```

# Ranking for all stores:

```python
query = """
WITH ranked AS (
  SELECT
    name,
    city,
    rating,
    review_count,
    score,
    RANK() OVER (
      ORDER BY score DESC, review_count DESC
    ) AS store_rank,
    COUNT(*) OVER () AS total_stores
  FROM store
)
SELECT *,
  ROUND(100.0 * store_rank / total_stores, 2) AS percent_rank
FROM ranked
WHERE name LIKE '%ICI tea drink%';

"""
pd.read_sql(query, conn)
```
